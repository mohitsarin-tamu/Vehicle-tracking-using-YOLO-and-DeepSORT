# Counting CV

```python
 

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def ccw(A, B, C):
    return (C.y - A.y) * (B.x - A.x) > (B.y - A.y) * (C.x - A.x)

def intersect(A, B, C, D):
    return ccw(A, C, D) != ccw(B, C, D) and ccw(A, B, C) != ccw(A, B, D)

def check_cross(checkline, traj_line):
    previous_x, previous_y = traj_line[1].x, traj_line[1].y
    next_x, next_y = traj_line[0].x, traj_line[0].y
    return intersect(Point(previous_x, previous_y), Point(next_x, next_y),
                     Point(checkline[0][0], checkline[0][1]), Point(checkline[1][0], checkline[1][1]))

```

```python
# Read the CSV file into a DataFrame
csv_file_path = 'bbox_coordinates.csv'
df = pd.read_csv(csv_file_path)
#changing frames to seconds 
df['Timestamp'] = pd.to_datetime((df['Frame']/15).astype(int), unit='s').dt.time # changing frame rate to that of video

# Convert 'Timestamp' to datetime and set as index
desired_date = date(2024, 10, 26)  # Set your desired date here
df['Timestamp'] = df['Timestamp'].apply(lambda t: datetime.combine(desired_date, t))
df.set_index('Timestamp', inplace=True)
df.head()
```

```python
# Organize data by Track ID
track_data = {}
# Initialize a dictionary to track crossed IDs for each counting line
crossed_track_ids = {i: set() for i in range(3)}  # assuming 3 counting lines for now

for _, entry in df.iterrows():
    track_id = entry['Track ID']
    if track_id not in track_data:
        track_data[track_id] = []
    track_data[track_id].append(entry)

# Access previous and next frames for each Track ID
for track_id, entries in track_data.items():
    # Skip processing if the Track ID has already crossed any of the lines
    if any(track_id in crossed_track_ids[line_idx] for line_idx in range(3)):
        continue

    for i, entry in enumerate(entries):
        current_frame = entry['Frame']
        current_midpoint = Point(entry['X_mid'], entry['Y_mid'])
        counting_lines = [
            [(727,182), (970,156)],
            [(1584,191), (1841,238)],
            [(158,373), (261,659)]
        ]
        if i > 0:
            previous_frame = entries[i - 1]['Frame']
            previous_midpoint = Point(entries[i - 1]['X_mid'], entries[i - 1]['Y_mid'])
            traj_line = [previous_midpoint, current_midpoint]

            # Check for crossings in each line
            for line_idx, counting_line in enumerate(counting_lines):
                if check_cross(counting_line, traj_line):
                    crossed_track_ids[line_idx].add(track_id)
                    break  # Stop checking once a crossing is detected for one line

        if i < len(entries) - 1:
            next_frame = entries[i + 1]['Frame']
            next_midpoint = Point(entries[i + 1]['X_mid'], entries[i + 1]['Y_mid'])
            traj_line = [current_midpoint, next_midpoint]

            # Check for crossings in each line
            for line_idx, counting_line in enumerate(counting_lines):
                if check_cross(counting_line, traj_line):
                    crossed_track_ids[line_idx].add(track_id)
                    break  # Stop checking once a crossing is detected for one line

# Count the number of crossings in 5-minute intervals for each line
crossings_count = {}
for line_idx in range(3):  # Assuming 3 lines
    crossings_count[line_idx] = df.groupby(pd.Grouper(freq='5T')).apply(
        lambda x: len(set(x['Track ID']).intersection(crossed_track_ids[line_idx]))
    )

# Print the number of crossings for each line
for line_idx, count in crossings_count.items():
    print(f"Total vehicles crossed line {line_idx + 1} every 5 minutes: {count}")
```

```python
# Assuming crossings_count is a list of three Series, one for each line
crossings_count_line1 = crossings_count[0]
crossings_count_line2 = crossings_count[1]
crossings_count_line3 = crossings_count[2]

# Ensure the index is in datetime format for all three lines
crossings_count_line1.index = pd.to_datetime(crossings_count_line1.index)
crossings_count_line2.index = pd.to_datetime(crossings_count_line2.index)
crossings_count_line3.index = pd.to_datetime(crossings_count_line3.index)

# Convert the index (timestamps) to relative time in hours for all three lines
start_time = crossings_count_line1.index[0]  # Assuming all lines have the same start time
relative_time_in_hours_line1 = (crossings_count_line1.index - start_time) / pd.Timedelta(hours=1)
relative_time_in_hours_line2 = (crossings_count_line2.index - start_time) / pd.Timedelta(hours=1)
relative_time_in_hours_line3 = (crossings_count_line3.index - start_time) / pd.Timedelta(hours=1)

# Generate hourly ticks for x-axis
hourly_ticks = [i for i in range(int(max(relative_time_in_hours_line1.max(), relative_time_in_hours_line2.max(), relative_time_in_hours_line3.max())) + 1)]  # Hourly intervals

# Plotting for Line 1
plt.figure(figsize=(10, 6))
plt.plot(relative_time_in_hours_line1, crossings_count_line1.values, marker='', linestyle='-', color='b')
plt.title('Vehicles Crossed Line 1 Every 5 Minutes')
plt.xlabel('Relative Time (Hours)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(hourly_ticks)
plt.savefig('vehicles_crossed_plot_line1.png')
plt.show()

# Plotting for Line 2
plt.figure(figsize=(10, 6))
plt.plot(relative_time_in_hours_line2, crossings_count_line2.values, marker='', linestyle='-', color='g')
plt.title('Vehicles Crossed Line 2 Every 5 Minutes')
plt.xlabel('Relative Time (Hours)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(hourly_ticks)
plt.savefig('vehicles_crossed_plot_line2.png')
plt.show()

# Plotting for Line 3
plt.figure(figsize=(10, 6))
plt.plot(relative_time_in_hours_line3, crossings_count_line3.values, marker='', linestyle='-', color='r')
plt.title('Vehicles Crossed Line 3 Every 5 Minutes')
plt.xlabel('Relative Time (Hours)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(hourly_ticks)
plt.savefig('vehicles_crossed_plot_line3.png')
plt.show()

# Convert the result for all three lines to a DataFrame with relative time
df_cv_car = pd.DataFrame({
    'Relative_Time (Hours)': relative_time_in_hours_line1,  # Assuming same relative time for all lines
    'Crossings_Count_Line1': crossings_count_line1.values
    # 'Crossings_Count_Line2': crossings_count_line2.values,
    # 'Crossings_Count_Line3': crossings_count_line3.values
})
```
